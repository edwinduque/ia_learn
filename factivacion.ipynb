```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(a):
    return 1/ (1 + np.exp(-a))
```

```python
x = np.linspace(10, -10, 100)
```

```python
x
```

```python
plt.plot(x, sigmoid(x))
```

```python
def step(x):
   return np.piecewise(x, [x<0.0, x>0.0],[0,1])
```

```python
plt.plot(x, step(x))
```

```python
def relu(x):
    return np.piecewise(x, [x<0.0,x>0.0],[0,lambda x:x])
```

```python
plt.plot(x, relu(x))
```

```python
def tan(a):
    return np.sin(a)/np.cos(a)
```

```python
plt.plot(x, tan(x))
```

```python
# y valor real
# y preduccion
def mse(y, y_hat, derivate=False):
    if derivate:
        return (y_hat-y)
    else:
        return np.mean((y_hat - y)** 2)
```

```python
real = np.array([0,0,1,1])
prediction = np.array([0.9, 0.5, 0.2, 0.0])
```

```python
mse(real,prediction)
```
